```python
x = np.array([10,20,30,50])

plt.pie(x)
plt.show() 
```

```python
x = np.array([10,20,30,50])
labelss = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"]
plt.pie(x,labels = labelss)
plt.show() 
```

```python
x = np.array([10,20,30,50])
labelss = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"]
plt.pie(x,labels = labelss,startangle = 90)
plt.show() 
```

```python
x = np.array([10,20,30,50])
labelss = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"]
explodee = [0.1,0.12,0.13,0.14]                           #Distance from center of circle
plt.pie(x,labels = labelss,startangle = 90,explode =explodee)
plt.show() 
```

```python
x = np.array([10,20,30,50])
labelss = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"]
explodee = [0.5,0,0,0]
plt.pie(x,labels = labelss,startangle = 90,explode =explodee,shadow = True)
plt.show() 
```

```python
x = np.array([10,20,30,50])
labelss = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"]
explodee = [0.5,0,0,0]
colorr = ['r','m','gold','g']
plt.pie(x,labels = labelss,startangle = 90,explode =explodee,shadow = True,colors = colorr)
plt.show() 
```

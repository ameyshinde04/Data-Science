```python
corr = data.corr()                       #corr doesnt perform calculation on string values
sns.heatmap(corr)               
```

Here in the below example heatmap is used it indicates that 
corr means correlation -
For  1 : It means perfect +ve correlation (as 1 feature increases other also increases)
For -1 : It means perfect -ve correlation (as 1 feature increases other get decreases)
For  0 : It means no correlation (these features are independent)

```python
numeric_data = data.select_dtypes(include=['number'])    # Select only numeric columns
corr = numeric_data.corr()                               # Compute the correlation matrix
sns.heatmap(corr,cmap='Blues', annot=True)
```

If you want multiple plots split by categories → Use FacetGrid
If you want to apply a function/type like histogram , scatterplot to each subplot → Use .map()

```python
import matplotlib.pyplot as plt
abc = sns.FacetGrid(data = tips, row ='smoker', col ='time') 

abc.map(plt.hist,'tip',color ='gold') 
```
